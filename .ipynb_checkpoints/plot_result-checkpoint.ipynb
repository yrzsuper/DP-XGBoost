```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gaussian_mechanism(delta, sigma, sensitivity):
    term1 = 2 * np.log(1.25/delta)
    term2 = sensitivity / sigma
    res = np.sqrt(term1) * term2
    return res
```

```python
def basic_composition(epochs, delta, sigma, sensitivity):
    history = []
    for epoch in np.arange(1, epochs+1):
        res = epoch * gaussian_mechanism(delta, sigma, sensitivity)
        history.append(res)
    return history
```

```python
def advanced_composition(epochs, delta, sigma, sensitivity, delta_prime):
    history = []
    for epoch in np.arange(1, epochs+1):
        epsilon = gaussian_mechanism(delta, sigma, sensitivity)
        term1 = 2 * epoch * np.log(2/delta_prime) * epsilon 
        term2 = epoch * epsilon * (np.e ** (epsilon) - 1)
        res = np.sqrt(term1) + term2
        history.append(res)
    return history
```

```python
advanced_res_1 = advanced_composition(epochs = 100,
                                      delta = 1e-5,
                                      sigma = 0.1,
                                      sensitivity = 1,
                                      delta_prime = 1e-1
                                     )
advanced_res_3 = advanced_composition(epochs = 100,
                                      delta = 1e-5,
                                      sigma = 0.3,
                                      sensitivity = 1,
                                      delta_prime = 1e-1
                                     )
advanced_res_5 = advanced_composition(epochs = 100,
                                      delta = 1e-5,
                                      sigma = 0.5,
                                      sensitivity = 1,
                                      delta_prime = 1e-1
                                     )
advanced_res_7 = advanced_composition(epochs = 100,
                                      delta = 1e-5,
                                      sigma = 0.7,
                                      sensitivity = 1,
                                      delta_prime = 1e-1
                                     )
```

```python
basic_res_1 = basic_composition(epochs = 100,
                  delta = 1e-5,
                  sigma = 0.1,
                  sensitivity = 1,
                 )
basic_res_3 = basic_composition(epochs = 100,
                  delta = 1e-5,
                  sigma = 0.3,
                  sensitivity = 1,
                 )
basic_res_5 = basic_composition(epochs = 100,
                  delta = 1e-5,
                  sigma = 0.5,
                  sensitivity = 1,
                 )
basic_res_7 = basic_composition(epochs = 100,
                  delta = 1e-5,
                  sigma = 0.7,
                  sensitivity = 1,
                 )
```

```python
plt.plot(basic_res_1, '-', label='scale = 0.1')
plt.plot(basic_res_3, '-', label='scale = 0.3')
plt.plot(basic_res_5, '-', label='scale = 0.5')
plt.plot(basic_res_7, '-', label='scale = 0.7')
plt.legend(fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.yscale('log')
plt.ylim(bottom=0, top=1e4)
plt.xlabel('iteration', fontsize=17)
plt.ylabel('privay', fontsize=17)
```

```python
hist_1 = []
delta = 1e-5
sensitivity = 1
scale_array = np.arange(0.1, 1.1, 0.1)
for sigma in scale_array:
    result = gaussian_mechanism(delta, sigma, sensitivity)
    hist_1.append(result)
```
